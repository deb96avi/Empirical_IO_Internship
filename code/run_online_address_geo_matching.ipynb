```python
import pandas as pd
import requests
input_df=pd.read_excel('optiques_200_centres_modified.xlsx')
input_df['shop_address_python'] = input_df['shop_address_1'].str.strip()
input_df['shop_address_python'] = input_df['shop_address_python'].str.replace(' ','+')
# API query for addresses have words separated by +. We are modifying the address string to that format and add commune name at end
# Eg:- 18 RUE DE L EGLISE 39110-->18+RUE+DE+L+EGLISE+39110+Aiglepierre
input_df["shop_address_python"] = input_df["shop_address_python"] +"+"+ input_df["name_commune_src_attr"]
```

```python
input_df.head(10)
```

```python
lat_col=[]
long_col=[]
missing_index=[]
lim=input_df.shape[0]
# Loop over each address. Try-exception clause included for 2 cases encountered:-
# 1. Address not present in input dataset
# 2. Address not available from API request

for i in range(lim):
  try:
    x = requests.get('https://api-adresse.data.gouv.fr/search/?q='+input_df['shop_address_python'][i])
    x=x.json()
  except:
    long_col.append('NA')
    lat_col.append('NA')
    missing_index.append(i)
    continue    
  try:
    long_col.append(x['features'][0]['geometry']['coordinates'][0])
  except:
    long_col.append('NA')
    missing_index.append(i)
  try:
    lat_col.append(x['features'][0]['geometry']['coordinates'][1])
  except:
    lat_col.append('NA')
```

```python
input_df['shop_lat']=lat_col
input_df['shop_long']=long_col
#[ZI+DE+LA+BARONNIE+73330+Le Pont-de-Beauvoisin,57++GRANDE+RUE+FRANCOIS+MITTERRAND+58400+La Charité-sur-Loire
 ]coordinates have been manually changed after download
```

```python
input_df.to_csv('optiques_200_centres_python_output.csv',encoding='ISO-8859-1')
```
